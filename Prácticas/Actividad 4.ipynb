```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../Analysis/team_stats_by_year.csv')
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
```

Gráfica de victorias por continente

```python
df_by_continent = df.groupby(['Continent', 'Year'])['Wins'].sum().reset_index()
df_by_continent.boxplot(by = 'Continent', figsize=(12,6))
plt.title('Victorias totales por continente')
plt.suptitle('')
plt.xlabel('Continentes')
plt.ylabel('Victorias totales')
plt.savefig("../img/boxplot_continent.png")
plt.show()
```

Evolución de victorias por continente

```python
df_by_continent = df.groupby(['Continent', 'Year'])['Wins'].mean().reset_index()
continents = df_by_continent['Continent'].unique()

for continent in continents:
    plt.figure(figsize=(16, 8))
    plt.title(f'Historial de victorias promedio por equipos de {continent}')
    continent_data = df_by_continent[df_by_continent['Continent'] == continent]
    sns.lineplot(x='Year', y='Wins', data=continent_data)
    plt.ylabel('Victorias promedio')
    plt.xlabel('Año')
    plt.savefig("../img/plot_continent_" + continent + ".png")
    plt.show()
```
